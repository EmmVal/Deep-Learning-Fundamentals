```python
import math
import random
import pandas as pd
import numpy as np
import pickle 
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick
import matplotlib as mpl
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.neighbors import NearestCentroid
from sklearn.metrics import silhouette_score
from sklearn.metrics import pairwise_distances
import warnings
```

```python
#Silence future warning to improve code performance in map charts with plotly
def hide_future_warning_plotly():
    warnings.filterwarnings(
        action="ignore",
        message=r"When grouping with a length-1 list-like, you will need to pass a length-1 tuple to get_group in a future version of pandas\. Pass `\(name,\)` instead of `name` to silence this warning\.",
        category=FutureWarning,
        module=r"plotly\.express\._core",
    )

hide_future_warning_plotly()
```

```python
def map_function_spread(db,lat,lon,level,color,size,type_map="open-street-map",zoom=4):
    
    df = db.copy()
    df['Hover_title'] = [f'Postcode: {int(lev)}' for lev in df[level]]
    df['log_crimes']=[np.round(np.log(1+x),2) for x in df[size]]
    df['Clusters']=[str(x) for x in df[color]]
    
    fig = px.scatter_mapbox(df, lat=lat, lon= lon, zoom=zoom,
                      mapbox_style= type_map
                    ,size='log_crimes',
                            size_max = 10,
                    width=1000,
                    height=850
                    ,opacity = .7
                    ,color = 'Clusters'
                    ,color_discrete_sequence=px.colors.qualitative.G10
                    ,hover_name='Hover_title')
    fig.show() 
```

```python
def closest_centroid(X_db,centroids,postcode_db):
    #centroids = model.cluster_centers_ for kmeans
    distances = pairwise_distances(centroids, X_db, metric='euclidean')
    ind = [np.argpartition(i,1)[:1] for i in distances]
    ind = [i[0] for i in ind]
    return pd.DataFrame({'Postcode Centroid':postcode_db.iloc[ind]}).reset_index(drop=True).reset_index().rename(columns={'index': 'Cluster'})
```

```python
def model_results(model,X_db,id,k):
    cluster_pred = model.fit_predict(X_db)
    s_score = silhouette_score(X_db, cluster_pred)

    model_id = id+str(k)
    model_ids.append(model_id)
    silhouette_scores.append(s_score)
    X_pred[model_id]=cluster_pred
```

```python
selected_var = ['lat','long','Priv_Rent','Count_All_crimes']
level = 'Postcode'
```

```python
seed = 0
```

### Load modeling dataset

```python
path = "Project Databases//"
with open(path+'Crime_modelling_db.pkl', 'rb') as f:
    Crime_modelling_db = pickle.load(f)
```

```python
# Shufflle data set
np.random.seed(seed)
Crime_modelling_db_s = Crime_modelling_db.sample(frac=1).reset_index(drop=True)
#print(Crime_modelling_db.shape)
Crime_modelling_db_s.head(5)
```

```python
X_db = Crime_modelling_db_s[selected_var]
```

```python
Postcodes = Crime_modelling_db_s[level]
```

### Variable standarization

```python
X_db_std = StandardScaler().fit_transform(X_db)
X_db_std = pd.DataFrame(X_db_std,columns=X_db.columns)
X_db_std.head(5)
```

```python
#Adding more weight to location variables
```

```python
X_db_std['lat'] = [2*x for x in X_db_std['lat']]
X_db_std['long'] = [2*x for x in X_db_std['long']]
```

### Modeling testing

```python
#Initial parameter setting

model_ids = []
silhouette_scores = []
k_max = 10 #no more than ten groups since adding many groups is not reccommended
k_min = 4 #at least 4 groups can be formed considering a map visualization of the data spread

X_pred = pd.DataFrame({level:Postcodes,
                      'lat':X_db['lat'],
                      'long':X_db['long'],
                       'n_crimes':X_db['Count_All_crimes']})

#Models for k-means
id = 'kmeans_k'
for k in list(range(k_min,k_max+1)):
    np.random.seed(seed)
    model = KMeans(n_clusters=k,n_init='auto',init='k-means++')
    model_results(model,X_db_std,id,k)

#Models for hierarchical standard clustering 
id = 'hierar_k'
for k in list(range(k_min,k_max+1)):
    np.random.seed(seed)
    model = AgglomerativeClustering(n_clusters=k)#,linkage='ward',metric='euclidean')
    model_results(model,X_db_std,id,k)

#Models for hierarchical linkage average and metric manhatan 
id = 'hierar_avg_manhattan'
for k in list(range(k_min,k_max+1)):
    np.random.seed(seed)
    model = AgglomerativeClustering(n_clusters=k,linkage='average',metric='manhattan')
    model_results(model,X_db_std,id,k)

#Models for hierarchical linkage average and metric cosine 
id = 'hierar_avg_cosine'
for k in list(range(k_min,k_max+1)):
    np.random.seed(seed)
    model = AgglomerativeClustering(n_clusters=k,linkage='average',metric='cosine')
    model_results(model,X_db_std,id,k)

#Models for hierarchical linkage complete and metric manhattan 
id = 'hierar_comp_manhattan'
for k in list(range(k_min,k_max+1)):
    np.random.seed(seed)
    model = AgglomerativeClustering(n_clusters=k,linkage='complete',metric='manhattan')
    model_results(model,X_db_std,id,k)

#Models for hierarchical linkage complete and metric cosine 
id = 'hierar_comp_cosine'
for k in list(range(k_min,k_max+1)):
    np.random.seed(seed)
    model = AgglomerativeClustering(n_clusters=k,linkage='complete',metric='cosine')
    model_results(model,X_db_std,id,k)
```

```python
final_model_results = pd.DataFrame({'model_id':model_ids,'silhouette_score':silhouette_scores})
```

```python
final_model_results.shape
```

### Reviewing top models

```python
final_model_results = final_model_results.sort_values(by='silhouette_score',ascending=False).reset_index(drop=True)
final_model_results.head(15)
```

Top models according to best silhouette_score:

# Hot spots

```python
model_final = 'hierar_k6'
```

```python
hot_spots_db = Crime_modelling_db_s.iloc[:,:5].reset_index(drop=True)
hot_spots_db['Cluster']=X_pred[model_final].reset_index(drop=True)
hot_spots_db['Coordinates'] = [(x,y) for x,y in zip(hot_spots_db['lat'],hot_spots_db['long'])]
```

```python
#Hot spots for clusters

hot_spots_db_vf = hot_spots_db.groupby(['Cluster']).agg({'Postcode': list,'Coordinates':list}).reset_index()
hot_spots_db_vf = hot_spots_db_vf.merge(hot_spots_db[['Cluster','Postcode']].groupby(['Cluster']).count().reset_index()
                                       ,how='left',on='Cluster')
hot_spots_db_vf = hot_spots_db_vf.merge(hot_spots_db[['Cluster','Priv_Rent','Count_All_crimes']].groupby(['Cluster']).sum().reset_index()
                                       ,how='left',on='Cluster')

hot_spots_db_vf = hot_spots_db_vf.rename({'Postcode_x': 'Postcodes', 'Postcode_y': 'N_postcodes','Priv_Rent':'Tot_priv_rent',
                                         'Count_All_crimes':'Tot_crimes'}, axis=1)

#compute the avg number of crimes by cluster considering their number of postcodes included in each cluster
hot_spots_db_vf['Avg_crimes'] = [round(x/y,0) for x,y in zip(hot_spots_db_vf['Tot_crimes'],hot_spots_db_vf['N_postcodes']) ]
#Sort values by the avg number of crimes in clusters
hot_spots_db_vf = hot_spots_db_vf.sort_values(by='Avg_crimes',ascending=False).reset_index(drop=True)

#Postcodes which are the centroids
clusters = X_pred[model_final]
clf = NearestCentroid()
clf.fit(X_db_std,clusters)
centroids = clf.centroids_
postcode_centroids = closest_centroid(X_db_std,centroids,Postcodes)
hot_spots_db_vf = hot_spots_db_vf.merge( postcode_centroids,how='left',on='Cluster')

hot_spots_db_vf
```

#### Conclusions

* As it can be seen and in accordance with the preliminary results in the data stage, cluster 4 consisting of the Adelaide CBD (postcode 500) is the riskiest zone. Autorithies need to pay special attention to this region because even though the territory is small, there is a considerable amount of crimes concentrated in this zone (mainly thefts and property damage).
* The next riskiest hotspot is cluster 0, which covers the surroundings of the CBD and includes the largest number of postal codes. Authorities can allocate resources to this area considering the proximity to the city center.
* Cluster 5 is third riskiest group covering a small region in the northest part of SA. Thus, police can allocate few rousrces focused on monitoring this small area.
* Finally there are three groups remaining located to the north, east and west sides of Adelaide. Authorithies can implement police stations in the heart of each of these regions to prevent future crimes.
* The postcodes identified as cluster's centroids can be proposed as locations to implement police stations in order to prevent future crimes in the respective cluster. They consist on suitable locations since their computation was based on postcodes' coordinates, number of private rentals and number of crimes.
