```python
import math
import pandas as pd
import numpy as np
import pickle 
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick
import matplotlib as mpl
from scipy.stats import pearsonr
```

```python
def scatter_plot(df,var_x,var_y):
    X = df[var_x]
    Y = df[var_y]
    label_x = cat_Var[var_x]['Label']
    label_y = cat_Var[var_y]['Label']
    
    plt.scatter(X, Y,c = 'firebrick',alpha=.5)
    plt.title(f"{label_y} by {label_x}")
    plt.xlabel(label_x)
    plt.ylabel('Number of crimes')
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    
    plt.savefig('Images//'+f"scatter_{var_x}.png",dpi=200)
    plt.show()

    corr,p_val = pearsonr(X,Y)
    print('Corr: ', round(corr,4))
    print('P-val: ', round(p_val,4))
```

```python
def box_plot_single(df,var_x):
    X = df[var_x]
    label_x = cat_Var[var_x]['Label']

    sns.set_theme(style="white")
    ax = sns.boxplot(y=df[var_x],color=".8",linecolor="#137", linewidth=.75)
    ax.set(ylabel=label_x)
    ax.set_title(f"Box Plot distribution for {label_x}")

    plt.savefig('Images//'+f"box_{var_x}.png",dpi=200)
    plt.show()
```

```python
def categoric_distribution_plot(db,variable,count,prop,title,var_cat = False):

    y = db[prop] #prop
    x = db.index #xvalues
    if var_cat:
        labels = [cat_Var[var]['Label'] for var in db[variable]] #labels for xvalues
    else:
        labels = [str(var) for var in db[variable]]
    counts = db[count].astype(int) #cases

    fig, ax1 = plt.subplots()
    fig.set_size_inches(8, 5)
    ax1.bar(x,counts,color='#F4F6F7',edgecolor = "darkgrey")
    ax1.set_ylabel('Number of crimes',fontdict={'family':'arial','color':'black','size':12})
    ax1.set_xticks(x)
    ax1.set_xticklabels(labels,size=10)
    ax1.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
    ax1.spines[['right', 'top']].set_visible(False)
    mpl.rcParams['axes.spines.top'] = False
    for i in range(len(counts)):
        ax1.annotate(str(f'{counts[i]:,}'), xy=(x[i],0), ha='center', va='bottom',textcoords="offset points",size=10,xytext=(0,2))
    ax2 = ax1.twinx()
    ax2.spines['right'].set_color('none') 
    ax2.tick_params( colors='none') 
    ax2.set_ylabel('% Crimes',fontdict={'family':'arial','color':'none','size':12}) #ylabel
    rot_set_xlabel = 5 #argument for xlabels
    ax2.set_xticklabels(labels,rotation=rot_set_xlabel,size=5) #xlabes
    ax2.yaxis.set_major_formatter(mtick.PercentFormatter(1,decimals=0)) #y axis as %
    for x,y in zip(x,y): #data values
        label = "{:.1%}".format(y)
        ax2.annotate(label,(x,y),textcoords="offset points",
                     xytext=(0,20),
                     ha='center',size=10,color='#030764')
    plt.title(title,size=13) #chart title
    plt.savefig('Images//'+f"{title}.png",dpi=200)
    plt.show()
```

```python
def corr_function(db):
    corr_df = db.corr().round(2)
    max_corr = 0.4
    mask = np.zeros_like(corr_df)
    mask[np.triu_indices_from(mask)] =  1

    with sns.axes_style("white"):
        f,ax = plt.subplots(figsize=(5,5))
        ax = sns.heatmap(corr_df,vmax=max_corr,mask=mask,square=True,annot=True,cmap='YlGnBu')
```

```python
cat_Var = {'Postcode':{'Label':'SA Postcode','Type':'Numeric','X_var':'No'},
            'lat':{'Label':'Latitude coordinate','Type':'Numeric','X_var':'Yes'},
            'long':{'Label':'Longitude coordinate','Type':'Numeric','X_var':'Yes'},
            'Area_km2':{'Label':'Postcode Area (km2)','Type':'Numeric','X_var':'Yes'},
            'Distance_center_km':{'Label':'Distance from Adelaide CBD (km)','Type':'Numeric','X_var':'Yes'},
            'Population_e':{'Label':'Population estimated','Type':'Numeric','X_var':'Yes'},
            'Priv_Rent':{'Label':'Number of private rentals','Type':'Numeric','X_var':'Yes'},
            'Count_Fraud':{'Label':'Fraud','Type':'Categoric','X_var':'Yes'},
            'Count_Murder':{'Label':'Murder','Type':'Categoric','X_var':'Yes'},
            'Count_Other_offences':{'Label':'Other offences','Type':'Categoric','X_var':'Yes'},
            'Count_Prop_Damage':{'Label':'Property Damage','Type':'Categoric','X_var':'Yes'},
            'Count_Theft':{'Label':'Theft','Type':'Categoric','X_var':'Yes'},
            'Count_All_crimes':{'Label':'All crimes (historically)','Type':'Numeric','X_var':'No'}}
```

## Scatter plots for numeric variables

```python
Y = 'Count_All_crimes'

numeric_X_var = [key for key in cat_Var.keys() if (cat_Var[key]['Type']=='Numeric' and cat_Var[key]['X_var']=='Yes')]
numeric_X_var
```

## Distribution for categoric variables (Type of Crime)

### Distribution for total crimes

```python
cat_X_var = [key for key in cat_Var.keys() if (cat_Var[key]['Type']=='Categoric')]
cat_X_var
```

```python
#Total Crime distribution
db_crime_gp = pd.DataFrame(Crime_datasets_vf[cat_X_var].sum()).reset_index().rename(columns={"index":'Type_crime',0:"Count"})
db_crime_gp['Total'] =  sum(db_crime_gp['Count'])
db_crime_gp['%Crime'] = db_crime_gp['Count']/db_crime_gp['Total'] 
db_crime_gp.sort_values(by='Count',ascending=False, inplace=True)
db_crime_gp.reset_index(drop=True,inplace=True)
db_crime_gp
```

```python
categoric_distribution_plot(db_crime_gp,'Type_crime','Count','%Crime','Historical crimes by Type in SA',True)
```

* Theft and Property Damage are the most common type of crimes in SA, accounting for almost 94% of the total crimes that have occurred historically.
* There are very few murders (only 737 historically) which is a good indicator

### Distribution top 10 postcodes with the most type of crimes

```python
db_crime_gp_postcode = Crime_datasets_vf.groupby(['Postcode'])[cat_X_var].sum().reset_index()

for variable in cat_X_var:
    
    db_sub = db_crime_gp_postcode[['Postcode']+[variable]].copy()
    db_sub['Total'] =  sum(db_sub[variable])
    db_sub['%Crime'] = db_sub[variable]/sum(db_sub[variable]) 
    db_sub = db_sub.sort_values(by=variable,ascending=False).reset_index(drop=True)
    db_sub = db_sub.iloc[:10].copy() #Top 10
    title_plot = f'Top 10 Postcodes with most {cat_Var[variable]["Label"]} cases'
    
    categoric_distribution_plot(db_sub,'Postcode',variable,'%Crime',title_plot)
```

* Postcodes 5000, 5108, 5112, 5113, 5114, 5162 and 5700 are generally in the top 10 of postcodes with the most crimes by each type. Hence, they are identified as possible hotspots

#### Grouping crimes

Considering that theft and property damage crimes both consist about 94% of the total sample plus the fact that in general the same postcodes appear in the top 10 for all crime types, then only the total number of crimes will be tested in the modeling stage instead of the specific type of crime

## Correlation among numeric variables

```python
corr_function(Crime_datasets_vf[numeric_X_var])
```

* Area and distance are highly correlated with latitude. Thus, we keep latitude to have the exact location of the hotspots
* latitude and longitude are negative correlated. We can keep them both to have the exact location of the hotspots
* Population and Private rent are highly correlated. We can keep Private rental as it has the best correlation with the target number of crimes
