#### Define functions neccesary to perform the analysis

```python
def pct_avg(df, column):
    '''Function to compute the Prevalence of Diabetes on df'''
    return sum(df[column])/df.shape[0]
```

```python
def train_val_test_sampling(df, pct_train, pct_val,seed):
    '''Function to split dataframes in 3 samples train, test and validation given a % for train and test '''
    pct_test = 1 -pct_train -pct_val
    df_aux = df[[df.columns[0]]] #select data for any column just to do the sampling

    #Shuffle and sampling data
    train_db, validate_db, test_db = np.split(df_aux.sample(frac=1, random_state=seed), 
                           [int(pct_train*len(df_aux)), int((pct_train+pct_val)*len(df_aux))])
    
    train_db['Sample'] = 'Train'
    validate_db['Sample'] = 'Val'
    test_db['Sample'] = 'Test'

    full_db = pd.concat([train_db, validate_db, test_db], axis=0)
    full_db = full_db[['Sample']]

    #Return a column for original df identifying in which sample belongs each observation
    return pd.concat([df_aux, full_db], axis=1)['Sample']
```

```python
def corr_function(db):
    '''Function to compute the matrix of correlations '''
    corr_df = db.corr().round(2)
    max_corr = 0.5
    mask = np.zeros_like(corr_df)
    mask[np.triu_indices_from(mask)] =  1

    with sns.axes_style("white"):
        f,ax = plt.subplots(figsize=(5,5))
        ax = sns.heatmap(corr_df,vmax=max_corr,mask=mask,square=True,annot=True,cmap='YlGnBu')
```

```python
def box_plot_single(df,var_x):
    '''Box plot for univariate numeric variables '''
    X = df[var_x]
    label_x = cat_Var[var_x]['Label']

    sns.set_theme(style="white")
    ax = sns.boxplot(y=df[var_x],color=".8",linecolor="#137", linewidth=.75)
    ax.set(ylabel=label_x)
    ax.set_title(f"Box Plot distribution for {label_x}")

    plt.savefig('./images/'+f"box_{var_x}.png",dpi=200)
    plt.show()
```

```python
def density_plot(df,var):
    '''Computes the histogram distribution of numeric variables'''
    sns.histplot(
    df[var], kde=True,
    stat="density", kde_kws=dict(cut=3))
    
    plt.title('Density distribution',size=12)
    plt.savefig('./images/'+f"Density_{var}.png",dpi=200)
    plt.show()
```

```python
def format_bin(db,variable):
    '''Formats the table used to compute a binary plot of numeric variables vs target'''
    upper_bound = db[variable]
    
    upper_bound = [x.right for x in upper_bound]
    upper_bound = ['{:.2f}'.format(x) for x in upper_bound]
    upper_bound = [str(x) for x in upper_bound]
    upper_bound[-1] = 'inf'
    
    db['UB'] = upper_bound
```

```python
def numeric_binning(db,target,bins,variable):
        '''Computes the table used to compute a binary plot for numeric variables vs target'''
        db_ax = db.copy()
        db_ax['metric_to_review'] = pd.qcut(db_ax[variable],bins, retbins=False, precision=4, duplicates='drop')

        db_summary = db_ax.groupby('metric_to_review')[target].aggregate(['count',np.mean]).reset_index()
        db_summary.columns = [variable,'Cases','DP'] #DP = Diabetes Prevalence
        
        format_bin(db_summary,variable)
        db_summary[variable] = [str(x) for x in db_summary[variable]]
        db_summary['Positive'] = db_summary['Cases']*db_summary['DP']
        db_summary['Positive'] = [round(x,0) for x in db_summary['Positive']]
        db_summary = db_summary[[variable,'Cases','Positive','DP','UB']]
        
        return db_summary
```

```python
def DP_distribution_numeric(db,target,bins,variable):

    '''Computes the binary plot for numeric variables vs target'''

    binning_db = numeric_binning(db,target,bins,variable)
    y = binning_db['DP'] 
    x = binning_db.index #xvalues

    labels = [str(x) for x in binning_db['UB']] #labels for x values
    counts = binning_db['Cases'] 

    fig, ax1 = plt.subplots()
    fig.set_size_inches(8, 5)

    ax1.bar(x,counts,color='#F4F6F7')
    ax1.set_ylabel('Total women',fontdict={'family':'arial','color':'black','size':11})
    ax1.set_xlabel(cat_Var[variable]['Label'],fontdict={'family':'arial','color':'black','size':11})
    ax1.set_xticks(x)
    ax1.set_xticklabels(labels,size=10)
    
    for i in range(len(counts)):
        ax1.annotate(str(f'{counts[i]:,}'), xy=(x[i],0), ha='center', va='bottom',textcoords="offset points",size=10,xytext=(0,0))

    ax2 = ax1.twinx()
    ax2.plot(y,marker='o',linestyle='-',color='darkred',label='DP',alpha=.7)  #plot
    ax2.set_ylabel('% DP',fontdict={'family':'arial','color':'black','size':11}) #ylabel
    #plt.ylim(0,.4) #y axis limit
    rot_set_xlabel = 0 #argument for xlabels
    #ax2.set_xticklabels(labels,rotation=rot_set_xlabel,size=1) #xlabes
    ax2.yaxis.set_major_formatter(mtick.PercentFormatter(1,decimals=0)) #y axis as %
    #ax2.legend()

    for x,y in zip(x,y): #data values
        label = "{:.0%}".format(y)
        ax2.annotate(label,(x,y),textcoords="offset points",
                     xytext=(0,5),
                     ha='center', size = 10)

    plt.title('Distribution of Diabetes Prevalence (DP)',size=12) #chart title
    fig.tight_layout()

    fig.savefig(f"./images/Bi_distr_{variable}.png",dpi=200)
    #plt.close()
    plt.show()
```

```python
def performance_metrics(y_actual, y_pred):
  '''Computes the performance metrics for the models tested'''
  conf = confusion_matrix(y_actual, y_pred)
  a_score = accuracy_score(y_actual,y_pred)
  TN = conf[0,0]
  FN = conf[1,0]
  FP = conf[0,1]
  TP = conf[1,1]
  FN_Rate = FN/(FN+TP) #False Negative Rate
  return a_score, FN_Rate
```

```python
def scores_baseline(n, p = .5):
    ''' Computes random scores for the baseline scenario'''
    np.random.seed(seed)
    return torch.tensor(bernoulli.rvs(p = p, size = n)).reshape(-1,1)
```

#### NN Model

```python
def set_seed(seed):
    '''Defines the initial seeds to make results replicable'''
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
```

```python
#Neural Network Class that inherits nn.Module
class SingleLayerPerceptron(nn.Module):
  def __init__(self, n_features):
    super().__init__() # instantiate our nn.Module
    # Single fully connected layer
    self.fc = nn.Linear(n_features, 1)

  def forward(self, x):
    x = self.fc(x)
    x = F.sigmoid(x) #Logistic function for activation function valid for binary results
    return x
```

```python
def train_test_NN(epochs, learnig_rate, momentum, 
                  X_train, X_test, y_train, y_test, p_threshold):
    '''Train the NN using the Single Layer Perceptron model gor a giving number of epochs, 
    learning rate, momentum train and test samples, and a p threshold to convert probabilities to lables 0,1'''

    model_id = f'M_LR[{learnig_rate}]_MO[{momentum}]'
    set_seed(seed)
    n_features = X_test.shape[1]
    
    # Create an instance of model
    model = SingleLayerPerceptron(n_features)
    # Set the criterion of model to measure the error, how far off the predictions are from the data
    criterion = nn.BCELoss()  # Binary Cross-Entropy Loss
    #Stochastic Gradient Descend with learning rate and momentum
    optimizer = optim.SGD(model.parameters(), lr = learnig_rate, momentum = momentum)
    
    # Train model
    
    for i in range(epochs):
      # Go forward and get a prediction
      y_pred = model(X_train) # Get predicted results
      # Measure the loss/error, gonna be high at first
      loss = criterion(y_pred, y_train) # predicted values vs the y_train
    
      # Do back propagation which takes the error rate of forward propagation and feed it back to find weights
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()
    
    print('Model: ',model_id,'Loss:',f'{loss.item():.4f}')
    
    # Evaluate Model
    with torch.no_grad():  # Turn off back propogation
      y_eval_probs = model(X_test) 
      loss_t = criterion(y_eval_probs, y_test)
      y_eval = (y_eval_probs > p_threshold).float() #Convert probs to class
    
    accuracy_test, FN_rate_test = performance_metrics(y_test, y_eval)

    model_test_results = pd.DataFrame({'ID':[model_id],
                              'FN_Rate':[FN_rate_test],
                              'Accuracy':[accuracy_test]})
    # Save Model
    torch.save(model.state_dict(),'./models/'+model_id+'.pt')
    return model_test_results
```

```python
#Defines the variables in the dataset and type
cat_Var = {
        'Pregnancies':  {'Label': 'Number of Pregnancies', 'Type' : 'Numeric'},
         'Glucose':  {'Label': 'Level of Glucose', 'Type' : 'Numeric'},
         'BloodPressure':  {'Label': 'Blood Pressure', 'Type' : 'Numeric'},
         'SkinThickness':  {'Label': 'Skin Thickness', 'Type' : 'Numeric'},
         'Insulin':  {'Label': 'Level of Insulin', 'Type' : 'Numeric'},
         'BMI':  {'Label': 'Body Mass Index (BMI)', 'Type' : 'Numeric'},
         'DPF':  {'Label': 'Diabetes Pedigree Function (DPF)', 'Type' : 'Numeric'},
         'Age':  {'Label': 'Age ', 'Type' : 'Numeric'},
         'Outcome':  {'Label': 'Diabetes outcome', 'Type' : 'Categoric'}
             }
```

It is not possible to have values of 0 in Glucose, BloodPressure, SkinThickness, Insulin and BMI. Thus, zero values will be changed to the median of each variable.

```python
#median imputation for variables with 0 invalid values
var_to_change = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for var in var_to_change:
    med = diabetes_df[var].median()
    diabetes_df[var] = np.where(diabetes_df[var]==0,med, diabetes_df[var])

diabetes_df.describe()
```

### Final Data Processing

```python
# Remove Pregnancies, SkinThickness and Insulin from analysis
X_db = diabetes_df[var_numeric]
X_db = X_db.drop(['Pregnancies','SkinThickness','Insulin'] ,axis=1)
print(X_db.columns)
```

```python
#Perform normal standarisation for numeric variables
X_db_std = StandardScaler().fit_transform(X_db)
X_db_std = pd.DataFrame(X_db_std,columns=X_db.columns)
```

```python
#Split processed data intro train, val and test data 

idx_train = diabetes_df['Sample'].isin(['Train'])
idx_val = diabetes_df['Sample'].isin(['Val'])
idx_test = diabetes_df['Sample'].isin(['Test'])

y_train = diabetes_df[idx_train][target]
y_val = diabetes_df[idx_val][target]
y_test = diabetes_df[idx_test][target]

X_train = X_db_std[idx_train].reset_index(drop=True)
X_val = X_db_std[idx_val].reset_index(drop=True)
X_test = X_db_std[idx_test].reset_index(drop=True)
```

## Modeling stage

```python
# Convert features to float tensors for NN training
X_train = torch.tensor(X_train.values, dtype=torch.float)
X_val = torch.tensor(X_val.values,dtype=torch.float)
X_test = torch.tensor(X_test.values,dtype=torch.float)

y_train = torch.tensor(y_train.values,dtype=torch.float).reshape(-1,1)
y_val = torch.tensor(y_val.values,dtype=torch.float).reshape(-1,1)
y_test = torch.tensor(y_test.values,dtype=torch.float).reshape(-1,1)
```

#### Baseline Scenario: Random choice 50/50

```python
#Baseline scenario - random choice
size = y_val.shape[0]
y_pred = scores_baseline(size)
accuracy_baseline, FN_rate_baseline = performance_metrics(y_val, y_pred)

model_results = pd.DataFrame({'ID':['Baseline'],
                              'FN_Rate':[FN_rate_baseline],
                             'Accuracy':[accuracy_baseline]})
model_results
```

#### Training and testing of NN with hyperparameters LR and Momentum

```python
epochs = 1000 #fixed to 1000 due to computational resources
p_threshold = 0.3 #Assuming a probability greater thatn 30% is Diabetes

#Hyperparameters to be improved with the validation sample:
learnig_rates = [.1, .01, .001, .0001] 
momentum = [x/10 for x in list(range(1,11,1))] #from .1 to 1 with step .1

combinations = list(product(learnig_rates, momentum))

for comb in combinations:
    learnig_rate = comb[0]
    momentum = comb[1]
    model_test_results = train_test_NN(epochs, learnig_rate, momentum,
                  X_train, X_val, y_train, y_val, p_threshold)
    
    model_results = pd.concat([model_results,
                               model_test_results],axis=0).reset_index(drop=True) #Append performance results for all models

print('Total models to be trained:', len(combinations))
```

## Final resutls

```python
#Select top 5 candidates with the greater balance between the least FN Rate and Highest Accuracy

model_results['Balance'] = [Accuracy/(FN_rate+1) for FN_rate, Accuracy in zip(model_results['FN_Rate'],
                                                                             model_results['Accuracy'])]

model_results = model_results.sort_values(by=['Balance'], ascending=[False])
model_results.to_excel('./models/model_results.xlsx',index=False)
model_results.head(5)
```

Hence, the best model is the NN with learning rate .001 and momentum .8, 
giving a False Negative Rate of 5.17% and Accuracy of 66.88%
